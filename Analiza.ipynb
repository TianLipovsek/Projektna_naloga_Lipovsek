Na prvem grafu sem analiziral povprečno število odigranih tekem po posameznih pozicijah.

```python
tekme_po_pozicijah = podatki.groupby("Pozicija").Stevilo_tekem.mean().plot()
tekme_po_pozicijah
```

Preveril sem, koliko igralcev je imelo dosegalo določeno povprečje točk na tekmo, pri čemer sem vzel le tiste, ki so dali vsaj 10 točk na tekmo. 

```python
tocke_vec_kot_deset = podatki[podatki.Tocke > 10]
```

```python
podtabela = tocke_vec_kot_deset[["Ime", "Priimek", "Pozicija", "Tocke"]].copy()
podtabela["zaokrozeno_povprecje_tock"] = tocke_vec_kot_deset["Tocke"].round()
podtabela.sort_values(by="Tocke", ascending = False, inplace = True)
igralci_po_tockah = podtabela.groupby(by = "zaokrozeno_povprecje_tock")[["Ime"]].count()
igralci_po_tockah
```

```python
igralci_po_tockah.plot.bar()
```

Nato sem vzel 20 najboljših strelcev, ki igrajo na poziciji PG (point guard) in pogledal njihov odstotek meta iz igre.

```python
met_PG = podatki[
    podatki.Pozicija == 'PG'
].round(0)

met_PG.head(20)
```

```python
met_PG.head(20).plot.bar(x='Ime', y='Met')
```
